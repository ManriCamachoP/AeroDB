```python
import pandas as pd
import numpy as np
```

```python
flights = pd.read_csv("flights.csv")
airlines = pd.read_csv('Airlines.csv')
airports = pd.read_csv("Airports_Data.csv")
```

```python
flights.drop(columns="Unnamed: 0", axis = 1, inplace = True)
flights.head()
```

```python
airlines.drop(columns="Unnamed: 0", axis = 1, inplace = True)
airlines.head()
```

```python
airports.drop(columns="Unnamed: 0", axis = 1, inplace = True)
airports.head()
```

```python
#Cleaning airlines dataframe

airlines.drop_duplicates(subset=['iaco'],inplace=True)
airlines.dropna(subset=['iaco'],inplace=True)

cols = list(airlines.columns)
airlines[cols] = airlines[cols].apply(lambda x: x.str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8'))
```

```python
airlines.head(15)
```

```python
airlines.to_csv('processed_airlines.csv')
```

```python
#To facilitate insertion of the data in the sql query - Airlines Table

for i in range(airlines.shape[0]):

    print(f"('{airlines.iloc[i][0]}','{airlines.iloc[i][1]}','{airlines.iloc[i][2]}','{airlines.iloc[i][3]}','{airlines.iloc[i][4]}'),")
```

```python
#Cleaning Airports dataframe

airports.drop_duplicates(subset=['ICAO Code'],inplace = True)
airports.dropna(inplace=True)
```

```python
airports.to_csv('processed_airports.csv')
```

```python
# To facilitate insertion of the data in the sql query - Airports Table

for i in range(airports.shape[0]):

    print(f"('{airports.iloc[i][0]}','{airports.iloc[i][1]}','{airports.iloc[i][2]}','{airports.iloc[i][3]}','{airports.iloc[i][4]}'),")
```

```python
#Cleaning the flights dataframe

#As we can see the csv file of flights have quotation marks on some columns so we need to clean that up

flights[['ICAO Code','Arrival Airport Code', 'Destination Airport Code','Flight Number']]= flights[['ICAO Code','Arrival Airport Code', 
         'Destination Airport Code','Flight Number']].apply(lambda x: x.str.replace("'", ""))
```

```python
#Now we need to work on the flight number because as we can compare with the airline dataframe we only need the first three letters

flights["iaco"] = flights["Flight Number"].apply(lambda x: str(x)[0:3])
```

```python
#Check if the arrival/destination Airport Code is in the Airports dataframe, if not replace all the values with 'nan' 

flights['Destination Airport Code'] = flights['Destination Airport Code'].apply(lambda x: 'nan' if x not in airports['ICAO Code'].values else x)
flights['Arrival Airport Code'] = flights['Arrival Airport Code'].apply(lambda x: 'nan' if x not in airports['ICAO Code'].values else x)



#Check if the IACO Code of the Flights are all in the Airline dataframe
flights['iaco'] = flights['iaco'].apply(lambda x: 'null' if x not in airlines['iaco'].values else x)
```

```python
flights.drop_duplicates(subset = "ICAO Code", inplace=True)
```
