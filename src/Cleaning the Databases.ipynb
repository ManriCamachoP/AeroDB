```python
import pandas as pd
```

```python
flights = pd.read_csv("flights.csv")
airlines = pd.read_csv('Airlines.csv')
airports = pd.read_csv("Airports_Data.csv")
```

```python
flights.drop(columns="Unnamed: 0", axis = 1, inplace = True)
flights.head()
```

```python
airlines.drop(columns="Unnamed: 0", axis = 1, inplace = True)
airlines.head()
```

```python
airports.drop(columns="Unnamed: 0", axis = 1, inplace = True)
airports.head()
```

```python
#As we can see the csv file of flights have quotation marks on some columns so we need to clean that up

flights[['ICAO Code','Arrival Airport Code', 'Destination Airport Code','Flight Number']]= flights[['ICAO Code','Arrival Airport Code', 
         'Destination Airport Code','Flight Number']].apply(lambda x: x.str.replace("'", ""))
```
